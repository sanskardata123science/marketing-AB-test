How frequently do we observe days with high number of conversions compared to days with low number of conversions?



````python
def conversion_category(conversion_col):
  category = []
  for conversion in df[conversion_col]:
    if conversion < 6:
      category.append('less than 6')
    elif 6<=conversion<10:
      category.append('6-10')
    elif 10<=conversion<15:
      category.append('10-15')
    else:
      category.append('more than 15')
  return category
````

````python
df['facebook_conversion_category'] = conversion_category('Facebook Ad Conversions')
df['adwords_conversion_category'] = conversion_category('AdWords Ad Conversions')
````

````python
facebook = pd.DataFrame(df['facebook_conversion_category'].value_counts().reset_index().rename(columns ={'facebook_conversion_category':'category'} ))
````

````python
adworks = pd.DataFrame(df['adwords_conversion_category'].value_counts().reset_index().rename(columns ={'adwords_conversion_category':'category'} ))
````

````python
category_df = pd.merge(facebook,adworks,on = 'category',how = 'outer').fillna(0)
````

````python
category_df
````

````python
category_df = category_df.iloc[[3,1,0,2]]
category_df
````

````python
import numpy as np
````

````python
X_axis = np.arange(len(category_df))
plt.figure(figsize = (15,6))
plt.bar(X_axis-0.2,category_df['count_x'],0.4,label = 'Facebook',color = '#03989E',linewidth = 1,edgecolor = 'k')
plt.bar(X_axis+0.2,category_df['count_y'],0.4,label = 'Adwords',color = '#A62372',linewidth = 1,edgecolor = 'k')
plt.xticks(X_axis,category_df['category'])
plt.xlabel("Conversion Category")
plt.ylabel("Number_of_days")
plt.title("frequency of daily conversion by conversion category",fontsize = 15)
plt.legend(fontsize = 15)
plt.show()
````

The data suggests Facebook had more frequent high conversion day than Adwords,which either had very low conversion rates (less than 6) or moderates ones(6-10)

There is significant variance in the number of high conversion days between two different campaigns.

The absence of any days with conversion between 10-15 and more than 15 in adwords indicates a need to review what strategies were changed or what external factors could have influence numbers

**How does Cost per Conversion(CPC) trend over time**

cost per conversion:This metrics is used to evaluate the cost effectiveness and profitablity of online advertising campaign.This metrics help marketers understand how much they are spending to obtain each conversion

````python
plt.figure(figsize=(8,5))
plt.title("Monthly cost per conversion")
monthly_df = df.groupby('month')[['Facebook Ad Conversions','Cost per Facebook Ad']].sum()
monthly_df['Cost per Conversion'] = monthly_df['Cost per Facebook Ad']/monthly_df['Facebook Ad Conversions']
plt.plot(month_names,monthly_df['Cost per Conversion'],'-o',color = '#A62373')
plt.show()
````

cpc trend over the 12-month period shows some fluctuations maintain a relatively stable ranges

may and november has lowest cpc value indicating potentially more cost effective advertising or higher conversion rates during these period

low cpc values in certain month could indicate period of higher advertising effectiveness or more favourable market condition
